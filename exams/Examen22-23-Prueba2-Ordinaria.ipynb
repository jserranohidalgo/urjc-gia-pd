__b) (0,5 puntos)__ Demuestra la siguiente deducción de la lógica proposicional clásica utilizando la correspondencia de Curry-Howard:

$\{ \neg q \rightarrow \neg p \} \vdash_c p \rightarrow q$

Para ello, se deberá asumir que se cumple la ley de la doble negación para la proposición $q$. Asimismo, se recomienda reutilizar en la implementación las demostraciones `dnI` y `contra` del apartado anterior.

```scala
def contraI[P, Q](p1: Not[Q] => Not[P])(DN: DoubleNegation): P => Q = 
    DN.apply[Q] compose contra(p1) compose DN.dnI[P]
```

```scala
def contraI[P, Q](p1: Not[Q] => Not[P])(DN: DoubleNegation): P => Q = 
    p => DN[Q](nq => p1(nq)(p))
```

```scala
type Not[P] = P => Nothing
```

```scala
trait DoubleNegation:
    def apply[P]: Not[Not[P]] => P
    def dnI[P]: P => Not[Not[P]] = 
        p => np => np(p)
```

```scala
def contra[P, Q](p1: P => Q): Not[Q] => Not[P] = 
    nq => p => nq(p1(p))
```

```scala
def compose[P, Q, R](g: Q => R, f: P => Q): P => R = 
    p => g(f(p))
```

```scala
def andThen[P, Q, R](f: P => Q, g: Q => R): P => R = 
    p => g(f(p))
```

```scala
def contraI[P, Q](p1: Not[Q] => Not[P])(DN: DoubleNegation): P => Q = 
    val dnI: P => Not[Not[P]] = DN.dnI[P]
    val con: Not[Not[P]] => Not[Not[Q]] = contra[Not[Q], Not[P]](p1)
    val dn: Not[Not[Q]] => Q = DN.apply[Q]
    val comp: P => Not[Not[Q]] = compose(con, dnI)
    compose(dn, comp) : (P => Q)
    // p => DN[Q](nq => p1(nq)(p))
    // dnQ compose contra(p1) compose dnI
    // Otras alternativas igualmente válidas:
    //    dnI andThen contra(p1) andThen dnQ
    //    p => dnQ(contra(p1)(dnI(p)))
    //    p => dnQ(nq => p1(nq)(p))
```

```scala
def contraI[P, Q](p1: Not[Q] => Not[P])(DN: DoubleNegation): P => Q = 
    val dnI: P => Not[Not[P]] = DN.dnI[P]
    val con: Not[Not[P]] => Not[Not[Q]] = contra[Not[Q], Not[P]](p1)
    val dn: Not[Not[Q]] => Q = DN.apply[Q]
    val comp: P => Not[Not[Q]] = andThen(dnI, con)
    andThen(comp, dn) : (P => Q)
    // p => DN[Q](nq => p1(nq)(p))
    // dnQ compose contra(p1) compose dnI
    // Otras alternativas igualmente válidas:
    //    dnI andThen contra(p1) andThen dnQ
    //    p => dnQ(contra(p1)(dnI(p)))
    //    p => dnQ(nq => p1(nq)(p))
```

```scala
def contraI[P, Q](p1: Not[Q] => Not[P])(DN: DoubleNegation): P => Q = 
    val dnI: P => Not[Not[P]] = DN.dnI[P]
    val con: Not[Not[P]] => Not[Not[Q]] = contra[Not[Q], Not[P]](p1)
    val dn: Not[Not[Q]] => Q = DN.apply[Q]
    val comp: P => Not[Not[Q]] = dnI.andThen(con)
    comp.andThen(dn) : (P => Q)
    // p => DN[Q](nq => p1(nq)(p))
    // dnQ compose contra(p1) compose dnI
    // Otras alternativas igualmente válidas:
    //    dnI andThen contra(p1) andThen dnQ
    //    p => dnQ(contra(p1)(dnI(p)))
    //    p => dnQ(nq => p1(nq)(p))
```

```scala
def contraI[P, Q](p1: Not[Q] => Not[P])(DN: DoubleNegation): P => Q = 
    val dnI: P => Not[Not[P]] = DN.dnI[P]
    val con: Not[Not[P]] => Not[Not[Q]] = contra[Not[Q], Not[P]](p1)
    val dn: Not[Not[Q]] => Q = DN.apply[Q]
    val comp: P => Not[Not[Q]] = con compose dnI
    dn compose comp : (P => Q)
    // p => DN[Q](nq => p1(nq)(p))
    // dnQ compose contra(p1) compose dnI
    // Otras alternativas igualmente válidas:
    //    dnI andThen contra(p1) andThen dnQ
    //    p => dnQ(contra(p1)(dnI(p)))
    //    p => dnQ(nq => p1(nq)(p))
```

__c) (1 punto)__ Demuestra la siguiente deducción de la lógica proposicional clásica utilizando la correspondencia de Curry-Howard:

$\{p \vee q, q \vee r \rightarrow s,\neg r \rightarrow \neg p\} \vdash_c s$

Para ello se deberá asumir que la ley de la doble negación se cumple para la proposición $r$. Asimismo, se recomienda reutilizar en la implementación la demostración `contraI` del apartado anterior.

```scala
def proof[P, Q, R, S](p1: Either[P, Q], p2: Either[Q, R] => S, p3: Not[R] => Not[P])(dnR: Not[Not[R]] => R): S = 
    p2(p1 match {
        case Left(p: P) => 
            Right(contraI(p3)(dnR)(p)) : Either[Q, R]
        case Right(q: Q) => 
            Left(q) : Either[Q, R]
    })
```

```scala
// Alternativamente, sin utilizar `contraI`
def proof[P, Q, R, S](p1: Either[P, Q], p2: Either[Q, R] => S, p3: Not[R] => Not[P])(dnR: Not[Not[R]] => R): S = 
    p2(p1 match {
        case Left(p) => 
            Right(dnR(p3(_)(p)))
        case Right(q) => 
            Left(q)
    })
```

__a) (1 punto)__ Implementa la función `exists` de manera recursiva (no final), de tal manera que el número de llamadas recursivas realizadas en una invocación sea mínimo.

```scala
def exists[A](l: List[A])(pred: A => Boolean): Boolean = 
    l match {
        case Nil => false
        case head :: tail if pred(head) => true
        case _ :: tail => exists(tail)(pred)
    }
```

__b) (1 punto)__ Implementa la función `exists` utilizando la función de orden superior `foldRight`.

```scala
def exists[A](l: List[A])(pred: A => Boolean): Boolean = 
    l.foldRight(false)((head, tailExists) => pred(head) || tailExists)
```

```scala
import scala.math.sqrt

def isPrime(n: Int): Boolean =
    if (n <= 3) 
        n > 1
    else if (n % 2 == 0 || n % 3 == 0) 
        false
    else {
        val limit: Int = sqrt(n).toInt
        for (i <- 5 to limit+1 by 6)
            if (n % i == 0 || n % (i+2) == 0) 
                return false
        true
    }
```

Implementa la función `isPrime` utilizando la función `exists` del apartado anterior:

```scala
import scala.math.sqrt

def isPrime(n: Int): Boolean =
    if (n <= 3) n > 1
    else if (n % 2 == 0 || n % 3 == 0) false
    else !exists((5 to sqrt(n).toInt+1 by 6).toList)(i => n % i == 0 || n % (i+2) == 0)
```

```scala
// Alternativamente, con un algoritmo menos eficiente y de manera idiomática:
def isPrime(n: Int): Boolean =
    !(2 to n-1).exists(n % _ == 0)
```

La función `reduce` devuelve el resultado de aplicar sucesivamente de izquierda a derecha una función binaria a todos los elementos de una lista. Si la lista es vacía devuelve el valor `None`. Por ejemplo:

```scala
class TestReduce(reduce: List[Int] => ((Int, Int) => Int) => Option[Int]) extends FlatSpec with Matchers{
    "reduce" should "work" in {
        reduce(List[Int]())(_+_) shouldBe None
        reduce(List[Int]())(_*_) shouldBe None
        reduce(List(1))(_+_) shouldBe Some(1)
        reduce(List(1))(_*_) shouldBe Some(1)
        reduce(List(1,2,3,4))(_+_) shouldBe Some(((1+2)+3)+4)
        reduce(List(1,2,3,4))(_*_) shouldBe Some(((1*2)*3)*4)
    }
}
```

__a) (1 punto)__ Implementa la función `reduce` mediante recursión final.

```scala
def reduce[A](l: List[A])(combine: (A, A) => A): Option[A] = {
    def aux(out: Option[A], list: List[A]): Option[A] = 
        (out, list) match {
            case (_, Nil) => out
            case (None, head::tail) => 
                aux(Some(head), tail)
            case (Some(out), head :: tail) => 
                aux(Some(combine(out, head)), tail)
        }
    
    aux(None, l)
}
```

```scala
// Alternativamente
def reduce[A](l: List[A])(combine: (A, A) => A): Option[A] = {
    def aux(out: A, list: List[A]): A = 
        list match {
            case Nil => out
            case head :: tail => 
                aux(combine(out, head), tail)
        }
    
    l match {
        case Nil => None
        case head :: tail => Some(aux(head, tail))
    }
}
```

```scala
run(new TestReduce(reduce))
```

__b) (1 punto)__ Implementa la función `reduce` mediante la función de orden superior `foldLeft`.

```scala
def reduce[A](l: List[A])(combine: (A, A) => A): Option[A] = 
    l.foldLeft(None: Option[A]){
        case (None, head) => 
            Some(head)
        case (Some(out), head) => 
            Some(combine(out, head))
    }
```

__c) (1 punto)__ Considérese la siguiente implementación imperativa de una función que calcula la suma de los cuadrados de los números en un rango $[a, b]$, i.e. $\Sigma_{i=a}^b i^2$:

```scala
import scala.math.pow

def sum(a: Int, b: Int): Int = {
    var out: Int = 0
    for (i <- a to b)
        out += pow(i,2).toInt
    out
}
```

La función satisface los siguientes tests:

```scala
class TestSum(sum: (Int, Int) => Int) extends FlatSpec with Matchers{
    "sum" should "work" in {
        sum(3,1) shouldBe 0 // if the range is empty, the result is 0
        sum(0,0) shouldBe 0
        sum(0,4) shouldBe 30 // sum of 0,1,4,9,16
        sum(1,10) shouldBe 385 // sum of 1,4,9,16,25,36,49,64,81,100
    }
}
```

```scala
run(new TestSum(sum))
```

Implementa la función `sum` mediante la función de orden superior `map` y la función `reduce` del apartado anterior.

```scala
import scala.math.pow

def sum(a: Int, b: Int): Int =
    reduce((a to b).toList.map(pow(_,2).toInt))(_+_) match {
        case None => 0
        case Some(a) => a
    }
```

```scala
// De manera más idiomática, haciendo uso de la librería estándar de Scala
import scala.math.pow

def sum(a: Int, b: Int): Int =
    (a to b).map(pow(_,2).toInt)
        .reduceOption(_+_)
        .getOrElse(0)
```
