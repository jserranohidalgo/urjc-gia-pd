````scala
// type Tree[A] = 1 + Tree[A] * A * Tree[A]

enum Tree[A]:
    case Empty()
    case Node(left: Tree[A], root: A, right: Tree[A])

object Tree:
    
    def void[A]: Tree[A] =
        Empty()
    
    def leaf[A](a: A): Node[A] =
        Node(Empty(), a, Empty())
    
    def right[A](a: A, tree: Tree[A]): Node[A] =
        Node(Empty(), a, tree)
    
    def left[A](tree: Tree[A], a: A): Node[A] =
        Node(tree, a, Empty())
    
    def node[A](left: Tree[A], a: A, right: Tree[A]): Node[A] =
        Node(left, a, right)
````

````scala
import Tree._
````

````scala
def foldTree[A, B](tree: Tree[A])(empty: B)(node: (B, A, B) => B): B =
    tree match
        case Empty() =>
            empty
        case Node(left, a, right) =>
            node(foldTree(left)(empty)(node),
                a,
                foldTree(right)(empty)(node))
````

Considera las siguientes dos funciones, ambas completamente implementadas:

- `tagParity` recorre una lista de enteros y separa los negativos (etiquetados como una cadena) de los no negativos (de los que devuelve su paridad).
- `measureWords` recorre una lista de cadenas y separa las de una sola letra (de las que devuelve ese carácter) de las más largas (de las que devuelve su longitud).

Obsérvese que ambas responden al mismo patrón estructural: recorren la lista, envían cada elemento a una de dos listas de salida y, al hacerlo, lo transforman. Las únicas diferencias son los tipos involucrados y el criterio que decide a qué lado va cada elemento (y en qué se transforma).

````scala
def tagParity(l: List[Int]): (List[String], List[Boolean]) =
    l match
        case Nil => (Nil, Nil)
        case head :: tail =>
            val (ss, bs) = tagParity(tail)
            if head < 0 then (head.toString :: ss, bs)
            else (ss, (head % 2 == 0) :: bs)

// tagParity(List(-1, 4, -3, 6)) == (List("-1", "-3"), List(true, true))
````

````scala
def measureWords(l: List[String]): (List[Char], List[Int]) =
    l match
        case Nil => (Nil, Nil)
        case head :: tail =>
            val (cs, is) = measureWords(tail)
            if head.length == 1 then (head.head :: cs, is)
            else (cs, head.length :: is)

// measureWords(List("a", "hello", "x", "world")) == (List('a', 'x'), List(5, 5))
````
